```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/dataset.csv')
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
df.head()
```

```python
# Check diseases
print("Total diseases:", df['Disease'].nunique())
print("\nDisease list:")
print(df['Disease'].value_counts())
```

```python
# Check symptoms
symptom_cols = [c for c in df.columns if 'Symptom' in c]
print("Symptom columns:", len(symptom_cols))

# Get all unique symptoms
all_symptoms = []
for col in symptom_cols:
    all_symptoms.extend(df[col].dropna().unique().tolist())

unique_symptoms = list(set([s.strip() for s in all_symptoms if str(s) != 'nan']))
print("Total unique symptoms:", len(unique_symptoms))
print("\nSample symptoms:", unique_symptoms[:20])
```

```python
# Plot top 20 diseases by count
plt.figure(figsize=(12, 6))
df['Disease'].value_counts().plot(kind='bar', color='steelblue')
plt.title('Disease Distribution in Dataset')
plt.xlabel('Disease')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.tight_layout()
plt.savefig('../data/disease_distribution.png')
plt.show()
print("Chart saved!")
```
